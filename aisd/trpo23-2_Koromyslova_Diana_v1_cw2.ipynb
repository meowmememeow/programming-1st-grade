1. Создайте класс Length (Длина), имеющий свойства

• value (значение),

• unit (единица измерения).

При изменении единицы измерения значение должно соответственно меняться. Например, при переходе от сантиметров к метрам значение должно уменьшаться в 100 раз. Допустимые значения свойства unit: ‘см’, ‘м’, ‘км’. Организуйте эту проверку. Продемонстрируйте работу с классом.

```python
#Решение 1.
class Length:
    def __init__(self, value, unit):
        # Инициализация свойств значения и единицы измерения
        self._value = value
        self._unit = unit

    def value(self):
        # Возвращает значение длины
        return self._value

    def unit(self):
        # Возвращает единицу измерения
        return self._unit

    def unit(self, new_unit):
        # Установка новой единицы измерения
        if new_unit not in ['см', 'м', 'км']:
            raise ValueError("Недопустимая единица измерения")

        # Конвертация значения в новую единицу измерения
        if self._unit == 'см':
            if new_unit == 'м':
                self._value /= 100
            elif new_unit == 'км':
                self._value /= 100000
        elif self._unit == 'м':
            if new_unit == 'см':
                self._value *= 100
            elif new_unit == 'км':
                self._value /= 1000
        elif self._unit == 'км':
            if new_unit == 'см':
                self._value *= 100000
            elif new_unit == 'м':
                self._value *= 1000

        # Установка новой единицы измерения
        self._unit = new_unit


# Пример использования
length = Length(100, 'см')
print("Изначальное значение:", length.value, length.unit)  # Изначальное значение: 100 см

# Изменяем единицу измерения
length.unit = 'м'
print("Значение после перевода в метры:", length.value, length.unit)  # Значение после перевода в метры: 1.0 м

# Изменяем единицу измерения ещё раз
length.unit = 'км'
print("Значение после перевода в километры:", length.value, length.unit)  # Значение после перевода в километры: 0.001 км

''' Вывод:
Изначальное значение: 100 см
Значение после перевода в метры: 1.0 м
Значение после перевода в километры: 0.001 км
'''
```

2. Создайте класс Ведомость, имеющий
атрибут класса:

• список_дисциплин (значением является список с названиями дисциплин);

свойства:

• дисциплина (при задании значения проверять наличие дисциплины в атрибуте список_дисциплин),

• группа;

методы:

• put – добавляет в ведомость информацию об оценке студента (фамилия, оценка – параметры метода). Для хранения данных внутри класса используйте словарь, в котором ключом является фамилия студента. Возможные оценки – «отлично», «хорошо», «удовл.», «неудовл.», «н/я»;

• get – возвращает оценку, полученную студентом (фамилия студента – параметр метода);

• change – изменяет оценку, полученную студентом (фамилия студента и новая оценка – параметры метода);

• del – удаляет информацию о студенте из ведомости (фамилия студента – параметр метода);

• result – возвращает кортеж из 5 чисел (количество каждого вида оценок в ведомости);

• __init__ – конструктор;

• __str__ – возвращает строку, содержащую заголовки (название экзамена, группа) и результаты экзамена в виде таблицы;

• count – возвращает количество студентов в ведомости;

• names – возвращает список фамилий, имеющихся в ведомости.

Продемонстрируйте работу с классом. 


```python
# Решение 2.
class Ведомость:

    список_дисциплин = ['Математика', 'Физика', 'История', 'Литература', 'Информатика']

    # конструктор;
    def __init__(self, дисциплина, группа):
        if дисциплина not in self.список_дисциплин:
            raise ValueError("Дисциплина не найдена в списке допустимых дисциплин")
        self.дисциплина = дисциплина
        self.группа = группа
        self.оценки = {}


    # добавляет в ведомость информацию об оценке студента (фамилия, оценка – параметры метода). 
    # Для хранения данных внутри класса используйте словарь, в котором ключом является фамилия студента. 
    #Возможные оценки – «отлично», «хорошо», «удовл.», «неудовл.», «н/я»;
    def put(self, фамилия, оценка):
        if оценка not in ['отлично', 'хорошо', 'удовл.', 'неудовл.', 'н/я']:
            raise ValueError("Недопустимая оценка")
        self.оценки[фамилия] = оценка


    # возвращает оценку, полученную студентом (фамилия студента – параметр метода);
    def get(self, фамилия):
        return self.оценки.get(фамилия, "Оценка не найдена")


    # изменяет оценку, полученную студентом (фамилия студента и новая оценка – параметры метода);
    def change(self, фамилия, новая_оценка):
        if новая_оценка not in ['отлично', 'хорошо', 'удовл.', 'неудовл.', 'н/я']:
            raise ValueError("Недопустимая оценка")
        if фамилия in self.оценки:
            self.оценки[фамилия] = новая_оценка
        else:
            print("Студент не найден в ведомости")


    # удаляет информацию о студенте из ведомости (фамилия студента – параметр метода);
    def delete(self, фамилия):
        if фамилия in self.оценки:
            del self.оценки[фамилия]
        else:
            print("Студент не найден в ведомости")


    # возвращает кортеж из 5 чисел (количество каждого вида оценок в ведомости);
    def result(self):
        отлично = list(self.оценки.values()).count('отлично')
        хорошо = list(self.оценки.values()).count('хорошо')
        удовл = list(self.оценки.values()).count('удовл.')
        неудовл = list(self.оценки.values()).count('неудовл.')
        н_я = list(self.оценки.values()).count('н/я')
        return (отлично, хорошо, удовл, неудовл, н_я)


    # возвращает строку, содержащую заголовки (название экзамена, группа) и результаты экзамена в виде таблицы;
    def __str__(self):
        result_str = f"Экзамен: {self.дисциплина}\nГруппа: {self.группа}\n\n"
        result_str += "{:<15} {:<15}\n".format("Фамилия", "Оценка")
        for фамилия, оценка in self.оценки.items():
            result_str += "{:<15} {:<15}\n".format(фамилия, оценка)
        return result_str


    # возвращает количество студентов в ведомости;
    def count(self):
        return len(self.оценки)


    # возвращает список фамилий, имеющихся в ведомости.
    def names(self):
        return list(self.оценки.keys())


# Пример использования
ведомость = Ведомость('Математика', 'Группа 1')

ведомость.put('Иванов', 'отлично')
ведомость.put('Петров', 'хорошо')
ведомость.put('Сидоров', 'удовл.')

print(ведомость)

print("Оценка студента Иванов:", ведомость.get('Иванов'))

ведомость.change('Иванов', 'неудовл.')
print(ведомость)

print("Количество студентов в ведомости:", ведомость.count())

print("Фамилии студентов в ведомости:", ведомость.names())

print("Результаты экзамена:", ведомость.result())


''' Вывод:
Экзамен: Математика
Группа: Группа 1

Фамилия         Оценка         
Иванов          отлично        
Петров          хорошо         
Сидоров         удовл.         

Оценка студента Иванов: отлично
Экзамен: Математика
Группа: Группа 1

Фамилия         Оценка         
Иванов          неудовл.       
Петров          хорошо         
Сидоров         удовл.         

Количество студентов в ведомости: 3
Фамилии студентов в ведомости: ['Иванов', 'Петров', 'Сидоров']
Результаты экзамена: (0, 1, 1, 1, 0)
'''
```

3. Создайте класс Заказ(Order), у которого есть свойства код_товара(code), цена(price), количество(count) и методы __init__ и __str__.

Создайте 2 класса-потомка: Опт(Opt) и Розница(Retail).
В этих классах создайте методы __init__, __str__.и сумма_заказа(summa), позволяющий узнать стоимость заказа.
Для опта стоимость единицы товара составляет 95% от цены, а при покупке более 500 штук – 90% цены.
В розницу стоимость единицы товара составляет 100% цены. Стоимость заказа равна произведению цены на количество. 

Продемонстрируйте работу с классами, создав необходимые объекты и обратившись к их свойствам и методам


```python
# Решение 3.
class Order:
    # добавляем свойства код товара, цена и количество в метод инит
    def __init__(self, code, price, count):
        self.code = code
        self.price = price
        self.count = count

    # возвращаем строку
    def __str__(self):
        return f"Товар: {self.code}, Цена: {self.price}, Количество: {self.count}"

# классы-потомки

class Opt(Order):
    def __init__(self, code, price, count):
        super().__init__(code, price, count)

    def __str__(self):
        return f"Оптовый заказ - {super().__str__()}"

    # позволякт узнать стоимость заказа
    # Для опта стоимость единицы товара составляет 95% от цены, а при покупке более 500 штук – 90% цены.
    # Стоимость заказа равна произведению цены на количество. 
    def summa(self):
        if self.count > 500:
            return self.price * self.count * 0.9
        else:
            return self.price * self.count * 0.95


class Retail(Order):
    def __init__(self, code, price, count):
        super().__init__(code, price, count)

    def __str__(self):
        return f"Розничный заказ - {super().__str__()}"

    # позволяет узнать стоимость заказа
    # В розницу стоимость единицы товара составляет 100% цены.
    # Стоимость заказа равна произведению цены на количество. 
    def summa(self):
        return self.price * self.count


# Пример использования
order_opt = Opt('123', 100, 600)
order_retail = Retail('456', 150, 10)

print(order_opt)
print("Сумма оптового заказа:", order_opt.summa())

print(order_retail)
print("Сумма розничного заказа:", order_retail.summa())

'''
Оптовый заказ - Товар: 123, Цена: 100, Количество: 600
Сумма оптового заказа: 54000.0
Розничный заказ - Товар: 456, Цена: 150, Количество: 10
Сумма розничного заказа: 1500
'''
```

Примеры, демонстрирующие ожидаемое поведение функции:

assert my_func_1([[[[[1, 2]]]]]) == [[[[[2, 1]]]]]

```python
# Решение 4.
def my_func_1(lst): # переворачивает список вложенных списков с произвольным уровнем вложенности.
    if not isinstance(lst, list):  # Проверяем, является ли элемент списком
        return lst

    reversed_lst = []  # Создаем пустой список для хранения перевернутых элементов
    for item in lst:
        reversed_lst.insert(0, my_func_1(item))  # Рекурсивно вызываем функцию для каждого элемента и добавляем его в начало списка
    return reversed_lst

# Проверка примеров
assert my_func_1(42) == 42
assert my_func_1([]) == []
assert my_func_1([1, 2]) == [2, 1]
assert my_func_1((1, 2)) == (1, 2)
assert my_func_1([[1, 2], [3, [4, 5]], 6]) == [6, [[5, 4], 3], [2, 1]]
assert my_func_1(['123', '456', '789', {10: [11, 12], 13: 14}, (15, 16)]) == [(15, 16), {10: [11, 12], 13: 14}, '789', '456', '123']
assert my_func_1([[[[[1, 2]]]]]) == [[[[[2, 1]]]]]

print(my_func_1(42))
print(my_func_1([]))
print(my_func_1([1, 2]))
print(my_func_1((1, 2)))
print(my_func_1([[1, 2], [3, [4, 5]], 6]))
print(my_func_1(['123', '456', '789', {10: [11, 12], 13: 14}, (15, 16)]))
print(my_func_1([[[[[1, 2]]]]]))

print("Все тесты пройдены успешно!")

'''
42
[]
[2, 1]
(1, 2)
[6, [[5, 4], 3], [2, 1]]
[(15, 16), {10: [11, 12], 13: 14}, '789', '456', '123']
[[[[[2, 1]]]]]
Все тесты пройдены успешно!
'''
```
